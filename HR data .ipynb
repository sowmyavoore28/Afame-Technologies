 # LABEL  ENCODING

```python
from sklearn.preprocessing import LabelEncoder

# Assuming cat_data is a DataFrame containing categorical data
# and df is your main DataFrame
encoder = LabelEncoder()

# Iterate over each column in cat_data
for col in cat_data.columns:
    # Check if the column exists in df
    if col in df.columns:
        # Encode the values of the column in df
        df[col] = encoder.fit_transform(df[col])
    else:
        print(f"Column '{col}' not found in DataFrame.")
```

# MODEL TRAINING

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=5)
```

# XGBOOST CLASSIFIER

```python
from xgboost import XGBClassifier
model=XGBClassifier()
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
ypred=model.predict(x_train)
accuracy_score(y_train,ypred)
```

```python
y_pred=model.predict(x_test)
```

# MODEL EVALUATION

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

# BALANCING THE DATASET

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE()
```

```python
x_smote, y_smote = smote.fit_resample(x_train,y_train)
```

```python
from collections import Counter
print("Actual Classes",Counter(y_train))
print("SMOTE Classes",Counter(y_smote))
```

# XGBOOST CLASSIFIER

```python
from xgboost import XGBClassifier
model1=XGBClassifier()
model1.fit(x_smote,y_smote)
```

```python
y_pred1=model1.predict(x_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
print(accuracy_score(y_test,y_pred1))
print(classification_report(y_test,y_pred1))
```
